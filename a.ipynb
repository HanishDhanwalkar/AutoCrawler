````python
def xpath_finder(htmlcode, instructions):
    prompt = f'''
        You're a web parser that is good at reading and understanding the HTML code and can give clear executable code on the brower.
        Please read the following HTML code, and then return an Xpath that can recognize the element in the HTML matching the instruction below.
        Instructions:
        {instructions}\n

        Here the HTML code:
        ```
        {htmlcode}
        ```
        If the HTML code doesn't contain the suitable information match the instruction, keep the xpath attrs blank.
        '''
    
    res = model.generate_content(prompt)
    return res.text

def selector_promter(htmlcode, instructions):
    prompt = f'''
        Please read the following HTML code, and then return an CSS selector that can recognize the element in the HTML matching the instruction below.
        Instructions:
        {instructions}

        Here the HTML code:
        ```
        {htmlcode}
        ```
        If the HTML code doesn't contain the suitable information match the instruction, keep blank string.
    '''
    res = model.generate_content(prompt)
    return res.text

def coder(htmlcode, instructions):
    selector = selector_promter(htmlcode, instructions)
    prompt = f'''
        Please read the following HTML code, and then return an python code that can recognize the element in the HTML matching the instruction below.
        Instructions:
            1. Code to scrape based on selector '{selector}'
            2. Import the package you need to use
            3. Define the extraction code as a function called 'extract_value', with one parameters 'HTML', so that we can input html with the parameters. The output of the function is the extraction result.
            4. Don't generate other parts of the code
        
        Here the HTML code:
        ```
        {htmlcode}
        ```
        If the HTML code doesn't contain the suitable information match the instruction, keep blank string.
    '''
    res = model.generate_content(prompt)
    return res.text
````

````python
# generated code
def extract_value(HTML):
    # Create a BeautifulSoup object from the HTML
    soup = bs4.BeautifulSoup(HTML, "html.parser")

    # Find the title element
    title = soup.find("title")

    # If the title element was found, return its text content
    if title:
        return title.text

    # If the title element was not found, return an empty string
    else:
        return ""
    
    
import os
folder_path = 'data\movie'
files =[]

with os.scandir(folder_path) as entries:
  for entry in entries:
    if entry.is_file():  
    #   print(entry.name)
        files.append(entry.name)
    
for file in files:
    with open(rf"{folder_path}\{file}", 'r', errors='ignore') as f:
        html = f.read()
    print(extract_value(html))
````
